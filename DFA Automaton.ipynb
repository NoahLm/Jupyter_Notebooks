```python
#DFA 
class automaton():
  q = []
  sigma = []
  delta = {}
  q0 = ""
  f = []
  qc = ""
  
  def initialize(self, q, sigma, delta, q0, f, qc):
    self.q = q
    self.sigma = sigma
    self.delta = delta
    self.q0 = q0
    self.f = f
    self.qc = qc
  
  def process_word(self,word):
        self.qc = self.q0
        word = str(word)
        word.replace(" ", "")
        key=0
        #checking all symbols in the word are part of sigma
        for i in range(0,len(word)):
            if(word[i] not in self.sigma):
                return "word is not accepted"
        #checking the word ends in a final state
        for i in word:
            mykey = (self.qc,i)
            print(mykey)
            #updating qc
            print(self.qc)
            self.qc = self.delta[mykey]
        if self.qc in self.f:
            print("word is accepted")
        else:
            print("word is not accepted")      
                
  def load_from_file(self, path):
    myfile = open(path, "r")
    myList = myfile.readlines()
    for a in range(len(myList)):
        myList[a] = myList[a].replace("{","").replace("}","").replace("\n","").replace("\t","").replace(" ","")
        myList[a] = myList[a].split(",")
    ql = []
    for a in myList[0]:
        ql.append(a)
    sigmal = []
    for a in myList[1]:
        sigmal.append(a)
    deltalp = []
    deltal = {}
    for a in myList[2]:
        deltalp.append(a) #dictionary is stored as a list so far
        q0l = myList[3]
    fl = []
    for a in myList[4]:
        fl.append(a)
    for i in deltalp:
        q_i, symbol , q_f = i.split('-')
        deltal[(q_i, symbol)] = q_f
    self.q = ql
    self.sigma = sigmal
    self.delta = deltal
    self.q0 = q0l[0]
    self.f = fl
    self.qc = q0l[0]

    #STR EN PROCESOOOOOOOOOOOOOOOO
  def __str__(self):
    ToPrint=("CURRENT AUTOMATON CONFIGURATION\n"+"\nInternal States="+str(self.q)+"\nAlphabet="+str(self.sigma)+
    "\nTransition Function"+str(self.delta)+"\nInitial State="+str(self.q0)+"\nSet of Final States="+str(self.f)+
    "\n\nThe Current State of the automaton is= "+str(self.qc))
    return print(ToPrint)

  def is_valid(q, sigma, delta, q0, f):
    import itertools as it
    #check if q0 is a single element and part of q
    q0list = []
    q0list.append(q0)
    if ((len(q0list) == 1) and (q0list[0] in q)) != True:
        raise Exception('NotValidConfigurationError')
        return False
    else:
        #make sure f is subset of q
        if (all(item in q for item in f)) != True:
            raise Exception('NotValidConfigurationError')
            return False
        else:
            #check if delta* comes properly from q and sigma combination
            max_comb = len(q) * len(sigma) #maximum number of combinations
            combinations = it.product(q,sigma)#stores the actual combinations
            for combination in combinations:
                #making sure all combinations are stored in delta and lead to an existing q
                if ((combination in delta == None) or (delta[combination] not in q)):
                    raise Exception('NotValidConfigurationError')
                    return False
                else: pass
            if max_comb != len(delta):
                raise Exception('NotValidConfigurationError')
                return False
            else:
                #if none of these conditionals rasied an error the automaton is valid
                return True
```
