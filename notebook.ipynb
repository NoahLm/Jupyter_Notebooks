```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
```

```python
sample_size = 10

dimension = 4
degree = 3

x = np.random.randn(sample_size, dimension)
```

```python
x
```

```python
pf = PolynomialFeatures(degree, include_bias=True)
pf.fit(x)
```

```python
pf.n_features_in_
```

```python
pf.n_output_features_
```

```python
pf.powers_
```

```python
pf.get_feature_names_out()
```

```python
pf.transform(x)
```
