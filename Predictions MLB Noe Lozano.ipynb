### FILL ALL DATA NULL WITH DATA INTERPOLATION FUNCTION

### RESHAPE THE TRAINING DF SO WE ONLY CONSIDER VALUABLE VARIABLES

```python
traindf = traindf.interpolate(method='linear')
```

```python
traindf.isna().sum()
```

```python
x = traindf.iloc[:,1:-1].to_numpy()
x = np.c_[np.ones(x.shape[0]), x]
y = traindf.iloc[:,-1].to_numpy().reshape(-1,1) 
```

## TRAINING AND SCORES

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
LR = LinearRegression(fit_intercept=False)
```

```python
LR.fit(x_train, y_train)
```

```python
LR.score(x_train, y_train)
```

```python
LR.score(x_test, y_test)
```

```python
y_pred = LR.predict(x_test)
MSE = mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(MSE)
```

## COEFFICIENTS EXPLANATION
In essence, if the coefficient is positive, it means that an increase in the corresponding feature is associated with an increase in the target variable. In this case, the varibles which has a positive coefficient are related to increase the number of runs.

In the other hand, if the coefficient is negative, it means that an increase in the corresponding feature is associated with a decrease in the target variable. In this case, runs are decreased.

And finally, it's important the magnitude of the coefficient. A larger magnitude indicates a stronger relationship between the feature and the target variable, while a smaller magnitude indicates a weaker relationship.

```python
for idx, col_name in enumerate(['feature_1']):
    coefs = LR.coef_[idx]
    print(f"{col_name}: {coefs}")
```
