```python
import pandas as pd
import numpy as np
```

```python
data_url = 'http://bit.ly/2cLzoxH'
gapminder = pd.read_csv(data_url)
print(gapminder)

df1 = gapminder[['continent', 'year','lifeExp']]

pd.pivot_table(df1, values='lifeExp', 
                     index=['year'], 
                     columns='continent')
print("\n\n",df1)

df1 = pd.pivot_table(df1, values='lifeExp',
                     index=['year'],
                     columns='continent',
                     aggfunc=[min,max])
print("\n\n",df1)
```
