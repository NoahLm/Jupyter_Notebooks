Regular Expression Processor that finds, replaces, and deletes matches in regular expressions.

```python
class RegExProc():
    def __init__(self, List):
        self.expressions = List

    def find(self, expression, string):
        return re.findall(expression, string)
  
    def replace(self, expression, string, value):
    # List of all matches to replace
        replace_match = self.find(expression, string)
        if("\n" in expression): #If there's a line break, which is not required to have a REGEX for it
            for match in replace_match:
                string = string.replace(match, value + " ")
        else:
      # Find matches to replace
            for match in replace_match:
                string = string.replace(match, value)
        return string

    def delete(self, expression, string):
    #Finding matches to delete
        delete_matches = self.find(expression, string)
    # Deleting the matches
        for match in delete_matches:
            string = string.replace(match, "")
            
    # Return the string without the matches
        return string

    def process(self, data):
        new_data = data
    
        #Check each expression and their required changes
        for expression in self.expressions:     
            if(expression[1] == "replace" and len(expression) == 3): #Because replace requires len of 3
                for i in range(0,len(data)):
                    new_data[i] = self.replace(expression[0], data[i], expression[2]) #using previous functions
            
            if(expression[1] == "delete"):
                for i in range(0,len(data)):
                    new_data[i] = self.delete(expression[0], data[i]) 
                    
            if(expression[1] == "find"):
                for i in range(0,len(data)):
                    new_data[i] = self.find(expression[0], data[i]) 

        return new_data #Processed strings
    

```

Individual testing of methods

```python
test_string_1 = "My email: 2009080@UPY.edu"
test_string_2 = "UPYLAMEJORUNI"
```

```python
ExpressionsTesting = RegExProc([(r"\w","replace","*"), (r"\b\w+@\w+.\w+","delete")])
```

```python
# Find emails
print(ExpressionsTesting.find(r"\b\w+@\w+.\w+", test_string_1))

# Replace emails by EMAIL
print(ExpressionsTesting.replace(r"\b\w+@\w+.\w+", test_string_1, "EMAIL"))

# Delete emails
print(ExpressionsTesting.delete(r"\b\w+@\w+.\w+", test_string_1))

# Find passwords
print('\n')
print(ExpressionsTesting.find(r"\w", test_string_2))

# Replace letters with *
print('\nMy password:', end=" ")
print(ExpressionsTesting.replace(r"\w", test_string_2, "*"))

# Delete letters
print('\nMy password:', end=" ")
print(ExpressionsTesting.delete(r"\w", test_string_2), "\n")
```
