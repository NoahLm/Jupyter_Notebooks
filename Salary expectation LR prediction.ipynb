Import a training library from sklearn.model_selection
Drop the Salary Column

```python
#Load your sklearn.model_selection train-split libary and then drop Salary
from sklearn.model_selection import train_test_split
x = df.drop('Salary',axis=1)
```

Print your new data

```python
#Print the Column left YearsExperience
x
```

Create data to train the model

```python
#Create variable y to train your model
y = pd.DataFrame(data=df['Salary'])
```

Print the first 5 rows of the y variable

```python
#Print y 5 first rows
y.head()
```

Create train and test sets using the library you've loaded before 

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

Import Linear Regression model to predict the data

```python
from sklearn.linear_model import LinearRegression

L=LinearRegression()
```

Fit your training data into the model

```python
#Remembre to use your training data to train the model
L.fit(xtrain,ytrain)
```

Using your test data. Predict the outcome

```python
y_pred=L.predict(xtest)
for num in y_pred:
  for res in num:
    print(res)
```

Print the scoring of the trained model with the test data

```python
print(L.score(xtest, ytest)) 
```
